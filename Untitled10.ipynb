```python
import kagglehub
import pandas as pd
import os

# Download dataset and read CSV
path = kagglehub.dataset_download("yasserh/housing-prices-dataset")
df = pd.read_csv(os.path.join(path, os.listdir(path)[0]))  # auto-load first CSV

# Preview
print(df.head())
```

```python
df.isnull().sum()
```

# No null value's , so need to worry about the null or missing Value here

```python
mapper ={'yes':1 ,'no':0}
df['mainroad'] = df['mainroad'].map(mapper)
df['basement'] = df['basement'].map(mapper)
df['hotwaterheating'] = df['hotwaterheating'].map(mapper)
df['airconditioning'] = df['airconditioning'].map(mapper)
df['prefarea'] = df['prefarea'].map(mapper)
df['guestroom'] = df['guestroom'].map(mapper)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Input (X) and Output (y)
X = df[['area']].values
y = df['price'].values

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predict
y_pred = model.predict(X)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))
print("RMSE:", rmse)
```
